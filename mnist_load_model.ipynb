```python
import numpy as np
import os
import tensorflow as tf
import matplotlib
#from tensorflow import keras
#from tensorflow.keras import layers, optimizers, datasets
#from matplotlib import pyplot as plt

tBatchSize = 128
# 数据读取和预处理
# 1. 读取训练集、测试集数据
# 2. 对数据进行预处理。X数据（像素灰度）进行归一化处理，一般处理到[0, 1]或[-1, 1]之间
#    对Y数据（分类标签）进行one-hot编码，去掉数字标签可能带来的大小关系
# 3. 使用tf.data.Dataset.from_tensor_slice((x, y)) 将三维的图片数据，按照第一个维度进行展开，及进行“打平”操作
# 4. 使用tf.data.Dataset.from_tensor_slice.batch 方法，设置批处理数据的大小
(x_in, y_in), (x_val, y_val) = tf.keras.datasets.mnist.load_data() #  如果没有from ... import ... 语句，需要一层层引用
x_in, x_val = tf.convert_to_tensor(x_in, dtype=tf.float32), tf.convert_to_tensor(x_val, dtype=tf.float32)
x_in, x_val = tf.reshape(x_in, [-1, 28*28]), tf.reshape(x_val, [-1, 28*28]) # shape=[60000, 784]
x_in, x_val = x_in / 255., x_val / 255.
# x_in = x_in.reshape(x_in.shape[0], 28*28) # np格式
y_in = tf.convert_to_tensor(y_in, dtype=tf.int32)
y_val = tf.convert_to_tensor(y_val, dtype=tf.int32)
y_in = tf.one_hot(y_in, 10) # tf 格式one-hot编码, 无论采用tf或者np格式，都需要对y进行one-hot编码
# y_in = (np.arange(10) == y_in[:, None]).astype(int)  # 整理输出one-hot编码的np格式数据
# y_val = (np.arange(10) == y_val[:, None]).astype(int)  


# 网络搭建
# step1 create model
model = tf.keras.Sequential([ # 三个非线性层的嵌套模型，包括两个隐藏层，一个输出层
        tf.keras.layers.Dense(256, activation='relu'), # 隐藏层1，用relu作为激活函数
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')]) # 输出层不用激活函数，输出节点数为10

# step2 compile model, build optimization method and loss function
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
# step3 train the model
# model.fit(x=x_in, y=y_in, epochs=1)
# model.build(input_shape=(None,28*28))
history = model.fit(x=x_in, y=y_in, batch_size=tBatchSize, epochs=10, 
             validation_split=0.3) # 返回值history记录了训练的相关过程, 输入数据可以是tf格式，或者np格式
```

```python
# step5 testing the model
loss, acc = model.evaluate(x_val, y_val)
```
