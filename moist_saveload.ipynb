```python
import os
import tensorflow as tf
from tensorflow import keras

# step1 加载mnist数据
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
x_test, x_train = x_test / 255.0, x_train / 255.0
x_test, x_train = x_test.reshape(-1, 28*28), x_train.reshape(-1, 28*28)

# step2 create model
def create_model():
    return keras.Sequential([
            keras.layers.Dense(256, activation='relu'),
            keras.layers.Dense(128, activation='relu'),
            keras.layers.Dense(64, activation='relu'),
            keras.layers.Dense(10, activation='softmax')
])
model = create_model()

# step3 configurate model
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=['accuracy'])


# step4 模型训练  训练一个epochs
model.fit(x=x_train,
          y=y_train,
          epochs=1,
          )

# step5 模型测试
loss, acc = model.evaluate(x_test, y_test)
print("train model, accuracy:{:5.2f}%".format(100 * acc))
```

```python
# step6 保存模型的权重和偏置
model.save_weights('my_save_weights.h5')
model.save('my_save.h5')
del model
```
